```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
x,y = make_moons(n_samples=1000, noise=0.2, random_state =0)
plt.plot(x[y==0, 0], x[y==0, 1], 'ob', alpha=1)
plt.plot(x[y==1, 0], x[y==1, 1], 'xr', alpha=1)
plt.legend(['0','1'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([
tf.keras.layers.Dense(1, input_dim=2, activation='sigmoid')
])

model.compile(tf.keras.optimizers.Adam(learning_rate=0.05), 'binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10, verbose=0)

print('Train Loss & Accuracy: ', model.evaluate(x_train, y_train))
print('Test Loss & Accuracy: ', model.evaluate(x_test, y_test))
```

```python
from mlxtend.plotting import plot_decision_regions


plot_decision_regions(X=x, y=y, clf=model, legend=1)
plt.xlabel('x', size=2)
plt.ylabel('y', size=2)
plt.title('Plot Decission Boundary', size=10)
plt.show()
```

```python
model1 = tf.keras.models.Sequential([
tf.keras.layers.Dense(1, input_dim=2, activation='sigmoid')
])

model1.compile(tf.keras.optimizers.Adam(learning_rate=0.05), 'binary_crossentropy', metrics=['accuracy'])
model1.fit(x_train, y_train, epochs=100, verbose=0)

print('Train Loss & Accuracy: ', model1.evaluate(x_train, y_train))
print('Test Loss & Accuracy: ', model1.evaluate(x_test, y_test))
```

```python
plot_decision_regions(X=x, y=y, clf=model1, legend=2)
plt.xlabel('x', size=5)
plt.ylabel('y', size=5)
plt.title('Plot Decission Boundary', size=10)
plt.show()
```
